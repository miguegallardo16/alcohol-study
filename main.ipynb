```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data\gapminder_alcohol.csv')
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df.info()
```

```python
my_order = df.sort_values('alcconsumption', ascending=False)
my_order.head()
```

```python
plt.figure(figsize=(35,12))
sns.barplot(data=my_order,x='country', y='alcconsumption')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(method ='spearman'),  annot=True)
```

```python
g = sns.pairplot(df)
g.map_offdiag(sns.kdeplot, levels=6, color=".2")
```
